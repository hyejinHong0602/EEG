# 전체 그래프 그려보기 : Trial1

```python
import pandas as pd
from datetime import datetime
 
df_data1 = pd.read_table('./data1_clean.csv',sep=',')
```

```python
df_data1
```

```python
# 그래프 그리기     
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(25,7))

# Line graph by seaborn
ax = sns.lineplot(data=df_data1[df_data1.columns[2:5]])
# ax = sns.lineplot(data=df[df.columns[-4:-1]][50:])
plt.title('EEG', fontsize=20)
plt.ylabel('amplitude', fontsize=14)

plt.xlabel('time(ms)', fontsize=14)
# plt.xticks([i for i in range(0,len(df), 1)])
# plt.xticks([i for i in range(50, len(df), 1)])
plt.legend(fontsize=12, loc='best')

# 세로 녹색 선 : 클릭 시점 표시해주기. 
for i in range(len(df_data1)) :
# for i in range(50,len(df)) :
    if df_data1['stimulus_time'][i] ==1 :
        min_ = [min(df_data1['Theta_AF8']),min(df_data1['Alpha_AF8']),min(df_data1['Beta_AF8'])]
        max_ = [max(df_data1['Theta_AF8']),max(df_data1['Alpha_AF8']),max(df_data1['Beta_AF8'])]
#         min_ = [min(df['mu8-12'][30:]),min(df['beta14-18'][50:]),min(df['beta36-40'][50:])]
#         max_ = [max(df['mu8-12'][50:]),max(df['beta14-18'][50:]),max(df['beta36-40'][50:])]
        plt.vlines(i,min(min_),max(max_), color="green")
plt.show()     
```

# df_sec로 그래프 그리기(1초별 데이터)

```python
import pandas as pd
from datetime import datetime
 
df_data1 = pd.read_excel('./df_sec.xlsx')
```

```python
df_data1 = df_data1.drop('Unnamed: 0', axis = 1)
```

```python
df_data1
```

```python
# 그래프 그리기     
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(25,7))

# Line graph by seaborn
ax = sns.lineplot(data=df_data1[df_data1.columns[1:4]][25:])
# ax = sns.lineplot(data=df[df.columns[-4:-1]][50:])
plt.title('EEG', fontsize=20)
plt.ylabel('amplitude', fontsize=14)

plt.xlabel('time(ms)', fontsize=14)
# plt.xticks([i for i in range(0,len(df), 1)])
plt.xticks([i for i in range(25, len(df_data1), 1)])
plt.legend(fontsize=12, loc='best')

# 세로 녹색 선 : 클릭 시점 표시해주기. 
for i in range(25,len(df_data1)) :
# for i in range(50,len(df)) :
    if df_data1['stimulus'][i] ==1 :
        min_ = [min(df_data1['Theta'][25:]),min(df_data1['Alpha'][25:]),min(df_data1['Beta'][25:])]
        max_ = [max(df_data1['Theta'][25:]),max(df_data1['Alpha'][25:]),max(df_data1['Beta'][25:])]
#         min_ = [min(df['mu8-12'][30:]),min(df['beta14-18'][50:]),min(df['beta36-40'][50:])]
#         max_ = [max(df['mu8-12'][50:]),max(df['beta14-18'][50:]),max(df['beta36-40'][50:])]
        plt.vlines(i,min(min_),max(max_), color="red")
plt.show()     
```
