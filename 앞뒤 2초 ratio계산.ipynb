```python
import pandas as pd
from datetime import datetime
 
df2= pd.read_excel('./df_sec2.xlsx')
df2
```

```python
df2 = df2.drop('Unnamed: 0', axis = 1)
```

```python
df2
```

```python
# # stimulus 시점의 행넘버 리스트에 넣어주기
stimulus_row_num = [] # click 시점의 행넘버
for k in range(len(df2)) : 
    if df2['stimulus'][k] == 1: # stimulus 시점의 행넘버 구하기
        stimulus_row_num.append(k)
print(stimulus_row_num)   
```

```python
# 빈 데이터프레임 생성.
df_ratio = pd.DataFrame(columns=['ratio_delta','ratio_theta','ratio_alpha','ratio_beta','ratio_gamma','stimulusOrNot'])
# 첫 클릭시점부터 마지막 클릭시점 이후 3초까지 루프 돌면서 증감율 구하기. 0123(클릭)4
for i in range(stimulus_row_num[0]-2, stimulus_row_num[-1]+3):
    stimulusOrNot = 0    
    if i+2 in stimulus_row_num :
        stimulusOrNot = 1
    ref_delta = df2[i:i+2].mean()[0]
    after_delta = df2[i+2:i+4].mean()[0]
    ratio_delta = (after_delta-ref_delta)/ref_delta
    
    ref_theta = df2[i:i+2].mean()[1]
    after_theta = df2[i+2:i+4].mean()[1]
    ratio_theta = (after_theta-ref_theta)/ref_theta
    
    ref_alpha = df2[i:i+2].mean()[2]
    after_alpha = df2[i+2:i+4].mean()[2]
    ratio_alpha = (after_alpha-ref_alpha)/ref_alpha
    
    ref_beta = df2[i:i+2].mean()[3]
    after_beta = df2[i+2:i+4].mean()[3]
    ratio_beta = (after_beta-ref_beta)/ref_beta
    
    ref_gamma = df2[i:i+2].mean()[4]
    after_gamma = df2[i+2:i+4].mean()[4]
    ratio_gamma = (after_gamma-ref_gamma)/ref_gamma
    
    
    df_ratio.loc[i+2]=[ratio_delta,ratio_theta,ratio_alpha,ratio_beta, ratio_gamma, stimulusOrNot]
```

```python
df_ratio.to_excel('./df_ratio2.xlsx')
```
