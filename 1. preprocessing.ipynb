```python
import pandas as pd
from datetime import datetime
 
stimulus = pd.read_table('./dataset/trial2.txt',sep='\n') 
stimulus.columns = ["time"]
stimulus['mark'] = 0
```

```python
time_0 = datetime.strptime('0:00:00.000000', "%H:%M:%S.%f") # 0초 만들기

for i in range(len(stimulus)) :
    for k in range(len(data)) :
        time_data = datetime.strptime(data['TimeStamp'][k], "%Y-%m-%d %H:%M:%S.%f")
        time_stimulus = datetime.strptime(stimulus['time'][i], "%Y-%m-%d %H:%M:%S.%f")
        time_interval= datetime.strptime(str(time_data - time_stimulus), "%H:%M:%S.%f")
        
        if time_interval > time_0 :
            data['stimulus'][i] = 1
            print(k)
            break
        
```
