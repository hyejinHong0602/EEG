```python
# 빈 데이터프레임 생성.
df_sec2 = pd.DataFrame(columns=['Delta','Theta','Alpha','Beta','Gamma','stimulus'])
```

```python
#전체 데이터 sampling rate인 256으로 잘라서 평균내서 1초당 데이터로 만들기.

delta_list=[]
theta_list=[]
alpha_list=[]
beta_list=[]
gamma_list=[]
stimulus_list=[]
for i in range(0,len(df_data2)//256):
    delta_list.append(df_data2[(i*256):((i+1)*256)].mean()[0])
    theta_list.append(df_data2[(i*256):((i+1)*256)].mean()[1])
    alpha_list.append(df_data2[(i*256):((i+1)*256)].mean()[2])
    beta_list.append(df_data2[(i*256):((i+1)*256)].mean()[3])
    gamma_list.append(df_data2[(i*256):((i+1)*256)].mean()[4])
    stimulus_list.append(df_data2[(i*256):((i+1)*256)].sum()[5])

df_sec2['Delta'] = delta_list
df_sec2['Theta'] = theta_list
df_sec2['Alpha'] = alpha_list
df_sec2['Beta'] = beta_list
df_sec2['Gamma'] = gamma_list
df_sec2['stimulus'] = stimulus_list
```
